```python
from typing import List, TypedDict
from langgraph.graph import StateGraph
import math
```

```python
class AgentState(TypedDict):
    name: str
    value: List[int]
    operation: str
    result: str
```

```python


def multiple_proccess(state: AgentState) -> AgentState:
    """ A node that process multiple imput """

    if state['operation'] == "+":
        state['result'] = f"Hi {state['name']}, your answer is {sum(state['value'])}"
    elif state['operation'] == "*":
        state['result'] = f"Hi {state['name']}, your answer is {math.prod(state['value'])}"
    else:
        state['result'] = "Invaild!"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("multi_proccessor", multiple_proccess)

graph.set_entry_point("multi_proccessor")
graph.set_entry_point("multi_proccessor")

app = graph.compile()
```

```python
result = app.invoke({"name": "Bob", "value": [1,2,3,4], "operation": "*"})
```
